```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [5,4,3,2,1]

plt.plot(x, y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [5,4,3,2,1]
c = 'red'

plt.plot(x, y, color = c)
plt.show()
```

```python
x = [10,20,30,40,50,60]
y = [60,50,40,30,20,10]
c = 'r'

plt.plot(x, y, color = c, marker = 'o')
plt.show()
```

```python
x = [10,20,30,40,50,60]
y = [60,50,40,30,20,10]
c = 'g'

plt.plot(x, marker = '>')
plt.plot(y, marker = 'o')
plt.plot(x, y, color = c, marker = '.')
plt.show()
```

# Bar Graph

```python
x = [10,20,30,40,50]
y = [50,40,30,20,10]

plt.bar(x, y)
plt.show()
```

```python
import numpy as np

x = np.array(['A','B','C','D','E'])
y = np.array([3,8,1,10,15])

plt.bar(x, y, color = 'g')
plt.show()
```

```python
x = np.array(['A','B','C','D','E'])
y = np.array([3,8,1,10,15])
c = ['yellow', 'green', 'blue', 'magenta', 'orange']

plt.bar(x, y, color = c)
plt.show()
```

```python
x = np.array(['A','B','C','D','E'])
y = np.array([3,8,1,10,15])
c = ['yellow', 'green', 'blue', 'magenta', 'orange']

plt.barh(x, y, color = c)
plt.show()
```

```python
x = np.array(['A','B','C','D','E'])
y = np.array([3,8,1,10,15])
c = ['yellow', 'green', 'blue', 'magenta', 'orange']

plt.bar(x, y, color = c, width = 0.5)
plt.show()
```

```python
x = np.array(['A','B','C','D','E'])
y = np.array([3,8,1,10,15])
c = ['yellow', 'green', 'blue', 'magenta', 'orange']

plt.barh(x, y, color = c, height = 0.5)
plt.show()
```

# Scatter Plot

```python
x = np.array(['A','B','C','D','E'])
y = np.array([3,8,1,10,15])
c = ['yellow', 'green', 'blue', 'magenta', 'orange']

plt.scatter(x, y, color = c)
plt.show()
```

```python
x = np.array(['A','B','C','D','E'])
y = np.array([3,8,1,10,15])
c = ['yellow', 'green', 'blue', 'magenta', 'orange']

plt.title('Scatter Plot')
plt.xlabel('Sections')
plt.ylabel('amount')
plt.scatter(x, y, color = c)
plt.show()
```

# Working with image

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

frame = r'logo.png'

#opening image using pil
image = Image.open(frame).convert('L')

#mapping image to gray scale
plt.imshow(image, cmap = 'coolwarm')

plt.title("Handshake")
plt.xlabel("Length")
plt.ylabel("Breadth")
plt.grid()
plt.show()
```

# Pie Chart

```python
x = [10,20,30,40,50]
y = ['English', 'Maths', 'Science', 'Social Science', 'General Knowledge']
c = ['yellow', 'green', 'orange', 'blue', 'pink']

plt.pie(x, labels = y, colors = c)
plt.legend()
plt.show()
```
